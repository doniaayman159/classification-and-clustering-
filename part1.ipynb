2. Initialize a function to plot all features within the dataset to visualize for outliers


```r
# Define a function to plot the features
plot_features <- function(data) {
  # Iterate over each column in the dataset
  for (column in names(data)) {
    # Check if the column contains numerical data
    if (is.numeric(data[[column]])) {
      # Create a box plot for numerical features
      boxplot(data[[column]], main = paste("Box Plot -", column), ylab = column)
    } else {
      # Create a bar plot for categorical features
      counts <- table(data[[column]])
      barplot(counts, main = paste("Bar Plot -", column), xlab = column, ylab = "Count")
    }
  }
}

# Call the function to plot the features
plot_features(data)
```

# 2) Tune the model using GridSearchCV

#Decision Tree

```r
library(rpart)

# Specify the formula for the decision tree
formula <- No.show ~ .

# Define the hyperparameters and their respective values to be tested
cp_values <- seq(0.01, 0.1, by = 0.01)
minsplit_values <- c(2, 5, 10)

# Initialize variables to store the best model and its associated accuracy
best_model <- NULL
best_accuracy <- 0

# Perform grid search for the decision tree
for (cp in cp_values) {
  for (minsplit in minsplit_values) {
    # Build the decision tree using rpart with current hyperparameter values
    dt_model <- rpart(formula, data = train_set, method = "class", control = rpart.control(cp = cp, minsplit = minsplit))

    # Predict using the current model
    dt_predictions <- predict(dt_model, newdata = test_set, type = "class")

    # Calculate accuracy
    cm <- confusionMatrix(dt_predictions, as.factor(test_set$No.show))
    accuracy <- cm$overall["Accuracy"]

    # Check if current model has higher accuracy than the best model so far
    if (accuracy > best_accuracy) {
      best_model <- dt_model
      best_accuracy <- accuracy
    }
  }
}

# Print the best model and its accuracy
print(best_model)
print(paste("Best Accuracy:", best_accuracy))
```

# NN

```r
# Assuming you have already loaded the keras, caret, and dplyr libraries

# Install the caret package if not already installed
# install.packages("caret")

# Assuming your train data is stored in the 'train_x1' and 'train_y1' variables, and test data in 'test_x1' and 'test_y1'

# Load the caret and dplyr libraries
library(caret)
library(dplyr)

# Define the function for creating and training the neural network model
create_model <- function(units, dropout_rate) {
  model <- keras_model_sequential() %>%
    layer_dense(units = units, input_shape = ncol(train_x1), activation = "relu") %>%
    layer_dropout(rate = dropout_rate) %>%
    layer_dense(units = 15, activation = "softmax")

  model %>% compile(
    loss = "categorical_crossentropy",
    optimizer = "adam",
    metrics = c("accuracy")
  )

  history <- model %>% fit(
    train_x1, train_y1,
    epochs = 50,
    batch_size = 32,
    validation_split = 0.2,
    verbose = 0
  )

  return(model)
}

# Specify the hyperparameters and their respective values to be tested
tuning_grid <- expand.grid(
  units = c(32, 64, 128),        # Example: Varying the number of units in the dense layer
  dropout_rate = c(0.2, 0.4, 0.6)  # Example: Varying the dropout rate
)

# Initialize variables to store the best model and its associated accuracy
best_model <- NULL
best_accuracy <- 0


```

```r
# Perform grid search for the neural network
for (i in 1:nrow(tuning_grid)) {
  # Build the neural network using the current hyperparameter values
  model <- create_model(units = tuning_grid$units[i], dropout_rate = tuning_grid$dropout_rate[i])

  # Evaluate the model on the test dataset
  loss_and_metrics <- model %>% evaluate(test_x1, test_y1, batch_size = 128)
  accuracy <- loss_and_metrics[1]

  # Check if current model has higher accuracy than the best model so far
  if (accuracy > best_accuracy) {
    best_model <- model
    best_accuracy <- accuracy
  }
}

# Print the best model and its accuracy
print(best_model)
print(paste("Best Accuracy:", best_accuracy))
```
